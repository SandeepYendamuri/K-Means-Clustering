```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
from sklearn.datasets import make_blobs
```

```python
#Create Data
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
```

```python
data
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c = data[1])  #centers as 4 so it's making 4 blobs 
```

```python
#Now lets create the clusters. 
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4,n_init=20,max_iter=100 )
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_ #centers are centroids 
```

```python
data[1]
```

```python
kmeans.labels_
```

```python
#To visualize the results

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

```python
K = 11
mse = np.zeros(K-1)
for n in range(1, K):
  kclusters = n
  kmeans = KMeans(n_clusters= kclusters, random_state=0).fit(data[0])
  mse[n-1] = kmeans.inertia_

plt.plot(range(1,K),mse)
plt.xlabel('K valule')
plt.ylabel('MSE')
```
